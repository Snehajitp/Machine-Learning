# Clustering

### Theoretical Questions:

1. What is unsupervised learning in the context of machine learning

Answer - Unsupervised learning is a type of machine learning where the model learns patterns from data without labeled outputs. The goal is to discover hidden structures such as clusters, groups, or relationships in the data. Common examples include clustering and dimensionality reduction.

2. How does K-Means clustering algorithm work

Answer  - K-Means works by dividing data into K clusters. It first randomly initializes K centroids, assigns each data point to the nearest centroid, then recalculates the centroids as the mean of assigned points. This process repeats until centroids stop changing.

3. Explain the concept of a dendrogram in hierarchical clustering

Answer -A dendrogram is a tree-like diagram that shows how clusters are merged or split at different distances. It helps visualize the clustering process and decide the optimal number of clusters by cutting the tree at a certain height.

4. What is the main difference between K-Means and Hierarchical Clustering

Answer - K-Means requires the number of clusters to be specified beforehand, while hierarchical clustering does not. K-Means is faster for large datasets, whereas hierarchical clustering provides a visual representation (dendrogram).

5. What are the advantages of DBSCAN over K-Means

Answer - DBSCAN can find clusters of any shape, does not require specifying the number of clusters, and can identify noise/outliers, which K-Means cannot handle well.

6. When would you use Silhouette Score in clustering

Answer - Silhouette Score is used to evaluate clustering quality by measuring how well data points fit within their clusters compared to other clusters. It is useful when comparing different clustering results or choosing the best number of clusters.

7. What are the limitations of Hierarchical Clustering

Answer - Hierarchical clustering is computationally expensive, does not scale well to large datasets, and once clusters are merged or split, the decision cannot be undone.

8.  Why is feature scaling important in clustering algorithms like K-Means

Answer - K-Means uses distance-based calculations, so features with larger values can dominate the results. Feature scaling ensures all features contribute equally to clustering.

9.  How does DBSCAN identify noise points

Answer - DBSCAN identifies noise as points that do not have enough neighboring points within a given distance (eps). These points are labeled as outliers.

10.  Define inertia in the context of K-Means

Answer - Inertia is the sum of squared distances between data points and their assigned cluster centroids. Lower inertia indicates tighter and more compact clusters.

11.  What is the elbow method in K-Means clustering

Answer - The elbow method plots inertia against the number of clusters. The point where the decrease in inertia slows down (forming an “elbow”) is considered the optimal number of clusters.

12. Describe the concept of "density" in DBSCAN

Answer - Density refers to how closely data points are packed together. DBSCAN forms clusters in regions with high point density and separates sparse regions as noise.

13.  Can hierarchical clustering be used on categorical data

Answer - Yes, hierarchical clustering can be used on categorical data by using appropriate distance measures like Hamming distance instead of Euclidean distance.


14. What does a negative Silhouette Score indicate


Answer - A negative Silhouette Score means data points are likely assigned to the wrong cluster and are closer to another cluster than their own.

15. Explain the term "linkage criteria" in hierarchical clustering

Answer - Linkage criteria define how the distance between clusters is measured, such as single, complete, average, or ward linkage.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities

Answer - K-Means assumes clusters are spherical and similar in size. When clusters vary in size or density, K-Means may assign points incorrectly.

17. What are the core parameters in DBSCAN, and how do they influence clustering

Answer - The two main parameters are:

* eps: radius for neighborhood search

* min_samples: minimum points required to form a dense region
They control cluster formation and noise detection.

18. How does K-Means++ improve upon standard K-Means initialization

Answer - K-Means++ selects initial centroids more carefully by spreading them apart, leading to faster convergence and better clustering results.

19. What is agglomerative clustering

Answer - Agglomerative clustering is a bottom-up hierarchical approach where each data point starts as its own cluster, and clusters are merged step by step.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer - Inertia only measures compactness, while Silhouette Score considers both cluster separation and cohesion, making it a more reliable evaluation metric.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200)
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []

for k in range(2,6):
    labels = KMeans(n_clusters=k, random_state=42).fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(range(2,6), scores)
plt.show()
```

31.  Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5)
kmeans = KMeans(n_clusters=3).fit(X)

xx, yy = np.meshgrid(
    np.linspace(X[:,0].min(), X[:,0].max(), 200),
    np.linspace(X[:,1].min(), X[:,1].max(), 200)
)

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.show()
```

40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
scatter plot.

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

digits = load_digits()
X = digits.data

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=labels)
ax.set_title("3D PCA + KMeans on Digits Dataset")
plt.show()
```
